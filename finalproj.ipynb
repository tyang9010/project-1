## 1. Data Preparation

Variables are divided into two categories: the expenditure variables and the household characteristics variables. For expenditure variables, percentages are used instead of the actual amount so it is comparable between different groups. Another advantage of percentages is to represent the contributions of expenditures for each group. Also the quarterly data are cleaned and annualized before usage.
For the household characteristics variables, they are separated into continuous variables (i.e, Number of Earners) and categorical variables (i.e, Age of Reference Person). Some of the categorical variables are recoded to be more interpretable.


```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import statsmodels.api as sm
import random
import sklearn
%pylab inline
pd.set_option('display.max_rows',9000)
pd.set_option('display.max_columns',9000)
figsize(18,12)
random.seed(1)
from pandas import Series,DataFrame
from matplotlib.mlab import PCA
from scipy.cluster import vq
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA,KernelPCA
from sklearn.lda import LDA
# Read 6 data files. doc121-doc131 are dataframes containg 2012 expenditure data for each quater.
# inquery : dataframe containing all the variable descriptions.
# fmly : query of 'FMLY' table variable descriptions

doc121 = pd.read_csv('fmli121x.csv')
doc122 = pd.read_csv('fmli122.csv')
doc123 = pd.read_csv('fmli123.csv')
doc124 = pd.read_csv('fmli124.csv')
doc131 = pd.read_csv('fmli131.csv')
inquery = pd.read_csv('Interview Query.csv')
fmly = inquery[inquery.TABLE=='FMLY'][['SECTION_DESCRIPTION','ORDER','VARIABLE_NAME','VARIABLE_DESCRIPTION']]
# Create variable list
varname1 = ['NEWID','CUID']
contkeep=['AS_COMP1', 'AS_COMP2', 'AS_COMP3','AS_COMP4', 
          'AS_COMP5', 'PERSLT18','PERSOT64','NO_EARNR','PRINEARN']
catkeep=['AGE_REF','REF_RACE','HIGH_EDU','BLS_URBN', 'CUTENURE', 'EARNCOMP','INCLASS']
varn6_5 = ['FOODPQ','ALCBEVPQ','HOUSPQ', 'APPARPQ','TRANSPQ','HEALTHPQ','ENTERTPQ','PERSCAPQ', 'READPQ','EDUCAPQ','TOBACCPQ','MISCPQ','CASHCOPQ','PERINSPQ','TOTEXPPQ']
varname = np.hstack((varname1,contkeep,catkeep,varn6_5))

# The following shows the expenditure variables and their descriptions. Note TOTEXPQ is equal to the sum of the remaining variables.
# fmly[fmly.VARIABLE_NAME.isin(varn6_5)][['VARIABLE_NAME','VARIABLE_DESCRIPTION']]
## To read selected columns from the dataset.
dat122 = doc122.loc[:,varname]
dat123 = doc123.loc[:,varname]
dat124 = doc124.loc[:,varname]
dat131 = doc131.loc[:,varname]
## To merge the data of four quarters.
data = pd.concat([dat122,dat123,dat124,dat131],keys=['one','two','three','four'])
# test here for misclassification
# Recode Catagorical data
data.AGE_REF[(data['AGE_REF']<=24)] = 1
data.AGE_REF[(data['AGE_REF']>=25) & (data['AGE_REF']<=34)] = 2
data.AGE_REF[(data['AGE_REF']>=35) & (data['AGE_REF']<=44)] = 3
data.AGE_REF[(data['AGE_REF']>=45) & (data['AGE_REF']<=54)] = 4
data.AGE_REF[(data['AGE_REF']>=55) & (data['AGE_REF']<=64)] = 5
data.AGE_REF[(data['AGE_REF']>=65) & (data['AGE_REF']<=74)] = 6
data.AGE_REF[data['AGE_REF']>=75] = 7
data.HIGH_EDU[data.HIGH_EDU<12]=1

# test here for input data set
# Annualize expenditure raw data
group = data.groupby('CUID')
expvar = np.hstack(('CUID',varn6_5))    ## Variable Used for the Expenditure part of the dataset.
expdata = data[expvar]     ## The expenditure part of the data
expitem = expdata.groupby('CUID').mean()*4                ## Annualized expenditure raw data
expitem2 = expitem.div(expitem.TOTEXPPQ,axis='index')     ## Convert every expenditure into percentage
expitem3 = expitem2.drop('TOTEXPPQ',axis=1)               ## Drop the TOTEXPPQ column from the percentage matrix
# Data cleaning for missing values
expitem3[expitem3==0] = np.nan
## expitem4 is used for PCA.
expitem4 = expitem3.dropna(how="all",axis=0)
expitem4 = expitem4.fillna(0)
expitem4.columns.values
expitem4.rename(columns={'FOODPQ':'Food', 'ALCBEVPQ':'Alcohol', 'HOUSPQ':'Housing', 'APPARPQ':'Apparel', 'TRANSPQ':'Transportation', 
                         'HEALTHPQ':'Health Care','ENTERTPQ':'Entertainment', 'PERSCAPQ':'Personal Care', 'READPQ':'Reading', 
                         'EDUCAPQ':'Education', 'TOBACCPQ':'Tobacco', 'MISCPQ':'Miscellaneous','CASHCOPQ':'Cash Contributions', 'PERINSPQ':'Personal Insurance'},inplace=True)

```

## 2. Data Analysis and Visualization

### 2.1 Clustering by Expenditure 

First or the most important step of data analysis is to identify the different consumer unit groups.  The K-Means result is plotted by the projecting the original data onto the plane spanned by the first two principal components from PCA. According to the results of several visualizations, the number of clusters is chosen to be three which performed the best. Notice that the PCA graphs may differ slightly across replications due to the unstable assignments of the first two principal components. 

```python
## expitem5 is standardized expitem4 (z score standardization)
expitem5 = expitem4.sub(expitem4.mean(1), axis=0).div(expitem4.std(1), axis=0)
np.random.seed(1)
km = KMeans(n_clusters=3, init='k-means++')
kmfit = km.fit(expitem5.values)
kmlabels = km.labels_
kmcluster_centers = km.cluster_centers_
kmlabels_unique = np.unique(kmlabels)
## Doing PCA on expenditure
#pcadata = expitem5.drop('ALCBEVPQ',axis=1)
np.random.seed(1)
pcadata = expitem5
pca = PCA(n_components=5)
exppca = pca.fit(pcadata.values).transform(pcadata.values)
## Plot for K-Means
plt.figure()
for c, i, target_name in zip("rbg", [0, 1, 2], kmlabels ):
    scatter = plt.scatter(exppca[kmlabels == i, 0], exppca[kmlabels == i, 1], c=c, label=target_name,s=100, alpha=0.3)

plt.xlabel("First Principal Component",fontsize=18)
plt.xticks(fontsize=14)
plt.ylabel("Second Principal Component",fontsize=18)
plt.yticks(fontsize=14)
plt.title('K-Means Clustering of Different Household Expenditure Groups',fontsize=25)
plt.text(-5.5,-5,'''
                Figure 1: Visualization of K-Means Clustering. The figure shows the distinct clusters by projecting data onto the plane
                spanned by the first two principal components. This figure shows our division of expenditure patterns is pretty reliable.''',fontsize=16)
plt.savefig('KMeans.png')
```

```python
## Need boxplots for all types of expenditures.
expitem4['Cluster']=''
expitem4.Cluster[(kmlabels==0)] = "Housing Dominant"
expitem4.Cluster[(kmlabels==1)] = "Food Dominant"
expitem4.Cluster[(kmlabels==2)] = "Transportation Dominant"
plt.rc('lines',linewidth=4)
for i in expitem4.columns.values[[0,2,4]]:
    expitem4.boxplot(column=i,by='Cluster',figsize=(8,6)).set_title("%s"%i,fontsize=20)
    plt.xticks(fontsize=12.5)
    plt.xlabel('')
    plt.ylabel("Percentage",fontsize=18)
    plt.savefig("%s.png"%i)
```

***Figure 2.* Boxplots for three typical types of expenditures**

Three boxplots of food, housing and transportation expenditures are used to investigate the different characteristics of these three groups. The group with the highest mean food expenditure percentage is labeled as “Food Dominant”. The same reasoning can be applied to “Housing Dominant” and “Transportation Dominant”.

### 2.3 Multinomial Logistic Regression Analysis

Multinomial Logisitc Regression is performed to evaluate which characteristic variables have what effects on a certain household, or say, what factors make a certain household fall into one of the three clusters.

The first table displays the regression coefficients.

The second table displays the marginal effects of all predictors.

The third table displays the comparison of marginal effects between three groups.

#### *Table 1.*  Multinomial Logistic Regression Coefficients

```python
datalogit = groupchar[contkeep]
for i in range(len(catkeep)):
    datalogit = datalogit.join(pd.get_dummies(groupchar[catkeep[i]], prefix=catkeep[i]).iloc[:,1:])
    
X=sm.add_constant(datalogit)
Y=expitem4.Cluster
model = sm.MNLogit(Y,X)
results = model.fit()
print results.summary()
```

#### *Table 2.*  Marginal Effects of Predictors

```python
# Warning to users: the code uses a massive amount of RAM
mdl_margeff = results.get_margeff()
print mdl_margeff.summary()
```


It is more straightforward to display the results in terms of marginal effect.


To interpret the table above, we mainly look at the "dy/dx" column. A positive number means that one unit increment of the predictor increases the probability of the household falling into that group by the corresponding "dy/dx" value. For catogorical variables, the result is interpreted by comparing to the baseline group(see next table).

For example, living in the suburb increases the probability of the household falling into Food Dominant group compared to living in the urban area. That means households in the suburb areas are more likely to spend more on food than those live in the cities.
